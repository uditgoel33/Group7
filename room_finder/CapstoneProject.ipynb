```python

##Identifying the room codes and cropping those images 
cropped = []
im.reset()
for i in range(0,len(full.curves)):
    if full.curves[i]['height'] < 10:
        box = full.curves[i]
        
        x0 = box['x0'] - 2
        
        top = box['top'] - 2
        bottom = box['bottom'] + 2
        x1 = box['x1'] + 2
        box =(x0, top, x1, bottom)
        cropped.append(full.crop(bbox=(box)))
        im.draw_rect((x0,top,x1,bottom))
        #print(full.extract_text(full.curves[i]))

im.save('RoomCodes.png',format='PNG')
```

```python
#Reading those images and trying to identify the room code 
for i,j in enumerate(cropped):
    image = j.to_image(resolution=144)
    path = '/users/uditgoel/Downloads/MastersDataScience/ImagesCapstone/image_' + str(i)  + '.png'
    image.save(path,format='PNG')
```
